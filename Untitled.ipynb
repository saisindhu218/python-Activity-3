```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances_argmin
```

```python
df = pd.read_csv("shopping_behavior_updated.csv")
df.head()
```

```python
df.columns
```

```python
X = df[['Age', 'Purchase Amount (USD)', 'Previous Purchases']]
X.head()
```

```python
X = X.dropna()
X.isnull().sum()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

labels_km = kmeans.labels_
centers_km = kmeans.cluster_centers_
```

```python
df['KMeans_Cluster'] = labels_km
df.head()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X_scaled[:,1], X_scaled[:,2], c=labels_km, alpha=0.6)
plt.scatter(centers_km[:,1], centers_km[:,2], marker='X', s=200)
plt.xlabel("Purchase Amount (scaled)")
plt.ylabel("Previous Purchases (scaled)")
plt.title("Standard K-Means Clustering")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_km, alpha=0.6)
plt.scatter(centers_km[:,0], centers_km[:,1], marker='X', s=200)
plt.xlabel("Age (scaled)")
plt.ylabel("Purchase Amount (scaled)")
plt.title("K-Means: Age vs Purchase Amount")
plt.show()
```
